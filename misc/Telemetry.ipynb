```python
import matplotlib.pyplot as plt
import mpld3
import numpy as np
import csv

mpld3.enable_notebook()
```

# Raw data ploting

```python
gyro_x = []
gyro_y = []
gyro_z = []
millis_time_gyro = []

with open('../tmp/tele_gyro.csv') as gyro_data:
    data = csv.DictReader(gyro_data, delimiter=';')
    for row in list(data):
        gyro_x.append(float(row['x']))
        gyro_y.append(float(row['y']))
        gyro_z.append(float(row['z']))
        millis_time_gyro.append(float(row['time(seconds)']))
```

```python
accel_x = []
accel_y = []
accel_z = []
millis_time_accel = []

with open('../tmp/tele_accel.csv') as accel_data:
    data = csv.DictReader(accel_data, delimiter=';')
    for row in list(data):
        accel_x.append(float(row['x']))
        accel_y.append(float(row['y']))
        accel_z.append(float(row['z']))
        millis_time_accel.append(float(row['time(seconds)']))
```

```python
g_f, ((g_ax1, a_ax1), (g_ax2, a_ax2), (g_ax3, a_ax3)) = plt.subplots(3, 2, sharex=True, figsize=(14,7))

g_ax1.set_title('MPU9255 gyroscope raw data.')
g_ax1.plot(millis_time_gyro, gyro_x, color='r')
g_ax2.plot(millis_time_gyro, gyro_y, color='g')
g_ax2.set_ylabel('dps')
g_ax3.plot(millis_time_gyro, gyro_z, color='b')
g_ax3.set_xlabel('seconds')

a_ax1.set_title('MPU9255 accelerometer raw data.')
a_ax1.plot(millis_time_accel, accel_x, color='r')
a_ax2.plot(millis_time_accel, accel_y, color='g')
a_ax2.set_ylabel('g')
a_ax3.plot(millis_time_accel, accel_z, color='b')
a_ax3.set_xlabel('seconds')

g_f.subplots_adjust(hspace=0)
plt.show()
```
